```python
import pandas as pd #manipular dados
import seaborn as srn #visualizar dados
import numpy as np #realizar operações em arrays multidimensionais
import matplotlib.pyplot as plt #gerar graficos
```

```python
data = pd.read_csv("dados.csv", sep=";")
data.head()
```

```python
#grafico de histograma do PIB
srn.histplot(data.PIB, kde=True, bins=10).set(title='PIB') #kde é a linha de densidade
```

```python
#grafico de histograma do PIB
srn.histplot(data.VALOREMPENHO, kde=True, bins=10).set(title='VALOR EMPENHO') #kde é a linha de densidade
```

```python
#agrupar 10 mmenores do PIB
agrupado = data.sort_values('PIB').head(10) #sort_values ordena pela coluna do PIB e .hea(10) mostra os 10 maiores valores
agrupado = agrupado.iloc[:, 1:3]
#pega so a coluna municipio e pib (nao pega a terceira coluna por ser excludente)
agrupado

#grafico (nao usar histograma pois agora quero ver relaçao de DUAS variaveis
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color='blue')
```

```python
#valores de empenho


agrupadov = data.sort_values('VALOREMPENHO').head(10)
agrupadov= agrupadov.iloc[:,1:4]
agrupadov

agrupadov.plot.bar(x='MUNICIPIO',y='VALOREMPENHO',color='red')
```

```python
data["PROPORCAO"] = data["VALOREMPENHO"]/data["PIB"]

prop = data.sort_values('PROPORCAO').head(10)
prop = prop.iloc[:, 1:5]
prop

prop.plot.bar(x='MUNICIPIO',y='PROPORCAO',color='green')

#esta indicando os de menores proporcoes, vamos analisar separadamente os de maiores proporcoes
```

```python
#maiores proporcoes

data["PROPORCAO"] = data["VALOREMPENHO"]/data["PIB"]

prop = data.sort_values('PROPORCAO', ascending=False).head(10)
prop = prop.iloc[:, 1:5]
prop

prop.plot.bar(x='MUNICIPIO',y='PROPORCAO',color='blue')

```

```python
#colocar os dois graficos lado a lado

# 10 menores proporções
prop_menores = data.sort_values('PROPORCAO').head(10)
prop_menores = prop_menores.iloc[:, 1:5]

# 10 maiores proporções
prop_maiores = data.sort_values('PROPORCAO', ascending=False).head(10)
prop_maiores = prop_maiores.iloc[:, 1:5]


#plotar lado a lado

fig, axes = plt.subplots(1,2,figsize=(14,6)) #uma linha com duas colunas

prop_menores.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color='green', ax=axes[0]).set(title="Menores proporcoes")

prop_maiores.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color='blue', ax=axes[1]).set(title="Maiores proporcoes")


plt.tight_layout() #para os graficos nao se sobreporem
plt.show()                                                                                     
```
